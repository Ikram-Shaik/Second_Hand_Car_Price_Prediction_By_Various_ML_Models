# <center>Datasets Cleaning and Merging</center>

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Datasets/dataset.csv")
```

```python
df1 = pd.read_csv("Datasets/cardekho_dataset.csv")
```

```python
df1.head()
```

```python
df.head()
```

## Dataset 1 Cleaning

```python
df.head()
```

```python
df = df[["Name","Year","Kilometers_Driven","Fuel_Type","Transmission","Mileage","Price"]]
```

```python
model_names = df1.model.unique()
model_names
```

```python
df["brand"] = df.Name.apply(lambda x : x.split()[0])
```

```python
df.head()
```

```python
df["model"] = df.Name.apply(lambda x : x.split()[1]+" "+x.split()[2] if x.split()[1]+" "+x.split()[2] in model_names
                                                                    else x.split()[1])
```

```python
df.head()
```

```python
df["mileage"] = df["Mileage"].apply(lambda x: str(x).split()[0] if pd.isna(x) == False else np.nan)
```

```python
df.head()
```

```python
df["price"] = df["Price"]*100000
```

```python
df["mileage"] = df["mileage"].astype('float')
df["price"] = df["price"].astype('int')
```

```python
df["mileage"].fillna(df.mileage.mean(),inplace=True)
```

```python
df.dtypes
```

```python
df.drop(['Name','Mileage','Price'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={"Year":"year","Kilometers_Driven":"km_driven","Fuel_Type":"fuel_type","Transmission":"transmission_type",
                   "price":"selling_price"},inplace=True)
```

```python
df = df[["brand","model","year","km_driven","fuel_type","transmission_type","mileage","selling_price"]]
```

```python
df.head()
```

## Dataset 2 Cleaning

```python
df1.head()
```

```python
df1.dtypes
```

```python
df1["year"] = 2023 - df1["vehicle_age"]
```

```python
df.head()
```

```python
df1 = df1[["brand","model","year","km_driven","fuel_type","transmission_type","mileage","selling_price"]]
```

```python
df1.head()
```

## Merging two dataframes

```python
final_df = pd.concat([df,df1])
```

```python
final_df.head()
```
